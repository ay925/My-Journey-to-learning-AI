```python
import pandas as pd

data={
    "Name":["Anupam","Anshu","Aryan","Ayush","Abhinav"],
    "Age":[17,None,17,18,None],
    "Salary":[54000,23000,45600,None,None]

}
df=pd.DataFrame(data)
# print(df)

print(df.isnull().sum())
drop_df=df.dropna()
print(drop_df)
df["Age"].fillna(df["Age"].mean(),inplace=True)
df["Salary"].fillna(df["Salary"].mean(),inplace=True)
print(df)
```
